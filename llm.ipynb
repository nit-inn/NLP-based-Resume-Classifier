```python
import os
import re
import pandas as pd
from docx import Document

# State and pincode mapping
PINCODE_STATE_MAP = {
    '11': 'Delhi', '12': 'Haryana', '13': 'Haryana', '14': 'Punjab', '15': 'Punjab', '16': 'Punjab',
    '17': 'Himachal Pradesh', '18': 'Jammu & Kashmir', '19': 'Jammu & Kashmir',
    '20': 'Uttar Pradesh', '21': 'Uttar Pradesh', '22': 'Uttar Pradesh', '23': 'Uttar Pradesh',
    '24': 'Uttar Pradesh', '25': 'Uttar Pradesh', '26': 'Uttar Pradesh', '27': 'Uttar Pradesh',
    '28': 'Uttar Pradesh',
    '30': 'Rajasthan', '31': 'Rajasthan', '32': 'Rajasthan', '33': 'Rajasthan', '34': 'Rajasthan',
    '36': 'Gujarat', '37': 'Gujarat', '38': 'Gujarat', '39': 'Gujarat',
    '40': 'Maharashtra', '41': 'Maharashtra', '42': 'Maharashtra', '43': 'Maharashtra', '44': 'Maharashtra',
    '45': 'Madhya Pradesh', '46': 'Madhya Pradesh', '47': 'Madhya Pradesh', '48': 'Madhya Pradesh',
    '49': 'Chhattisgarh',
    '50': 'Andhra Pradesh', '51': 'Andhra Pradesh', '52': 'Andhra Pradesh', '53': 'Andhra Pradesh',
    '56': 'Karnataka', '57': 'Karnataka', '58': 'Karnataka', '59': 'Karnataka',
    '60': 'Tamil Nadu', '61': 'Tamil Nadu', '62': 'Tamil Nadu', '63': 'Tamil Nadu', '64': 'Tamil Nadu',
    '67': 'Kerala', '68': 'Kerala', '69': 'Kerala',
    '70': 'West Bengal', '71': 'West Bengal', '72': 'West Bengal', '73': 'West Bengal', '74': 'West Bengal',
    '75': 'Odisha', '76': 'Odisha', '77': 'Odisha',
    '78': 'Assam', '79': 'Northeast States', '80': 'Bihar', '81': 'Bihar', '82': 'Bihar', '83': 'Bihar',
    '92': 'Jharkhand',
}

# State and native language mapping
STATE_NATIVE_LANGUAGES = {
    'Delhi': ['Hindi', 'English'],
    'Haryana': ['Hindi', 'Punjabi'],
    'Punjab': ['Punjabi', 'Hindi'],
    'Himachal Pradesh': ['Hindi', 'Pahari'],
    'Jammu & Kashmir': ['Kashmiri', 'Urdu'],
    'Uttar Pradesh': ['Hindi', 'Urdu'],
    'Rajasthan': ['Hindi', 'Rajasthani'],
    'Gujarat': ['Gujarati', 'Hindi'],
    'Maharashtra': ['Marathi', 'Hindi'],
    'Madhya Pradesh': ['Hindi', 'Bundeli'],
    'Chhattisgarh': ['Hindi', 'Chhattisgarhi'],
    'Andhra Pradesh': ['Telugu', 'Urdu'],
    'Karnataka': ['Kannada', 'Tulu'],
    'Tamil Nadu': ['Tamil', 'English'],
    'Kerala': ['Malayalam', 'Tamil'],
    'West Bengal': ['Bengali', 'Hindi'],
    'Odisha': ['Odia', 'Hindi'],
    'Assam': ['Assamese', 'Bengali'],
    'Northeast States': ['English', 'Local Languages'],
    'Bihar': ['Hindi', 'Maithili'],
    'Jharkhand': ['Hindi', 'Santali'],
}

# Function to extract text from a Word document
def extract_text_from_docx(file_path):
    try:
        doc = Document(file_path)
        full_text = "\n".join([p.text for p in doc.paragraphs])
        return full_text
    except Exception as e:
        print(f"Error processing file {file_path}: {e}")
        return ""

# Function to detect languages in the text
def detect_languages(text):
    language_list = [
        "English", "Spanish", "French", "German", "Italian", "Chinese", "Japanese", "Hindi",
        "Arabic", "Russian", "Portuguese", "Korean",
        "Assamese", "Bengali", "Bodo", "Dogri", "Gujarati", "Kannada", "Kashmiri", "Konkani",
        "Maithili", "Malayalam", "Manipuri", "Marathi", "Nepali", "Odia", "Punjabi", "Sanskrit",
        "Santali", "Sindhi", "Tamil", "Telugu", "Urdu"
    ]
    return [lang for lang in language_list if re.search(rf'\b{lang}\b', text, re.IGNORECASE)]

# Function to determine state from pincode
def get_state_from_pincode(pincode):
    pincode_prefix = pincode[:2]
    return PINCODE_STATE_MAP.get(pincode_prefix, "Unknown")

# Main function to process employee data
def process_employees(cv_folder, employee_csv, output_csv):
    # Read the employee data
    df = pd.read_csv(employee_csv, encoding='ISO-8859-1')

    language_sorted_results = []
    detected_languages_list = []

    for _, row in df.iterrows():
        emp_id = row['CandidateID']
        pincode = str(row['Branch Pincode'])
        state = get_state_from_pincode(pincode)

        # Get native languages for the state
        native_languages = STATE_NATIVE_LANGUAGES.get(state, [])

        # Locate CV file
        cv_filename = f"{emp_id} Resume.docx"
        cv_path = os.path.join(cv_folder, cv_filename)

        if not os.path.exists(cv_path):
            print(f"CV file not found: {cv_path}")
            language_sorted_results.append(0)
            detected_languages_list.append([])
            continue

        # Process the CV file
        cv_text = extract_text_from_docx(cv_path)
        cv_languages = detect_languages(cv_text)

        # Check if detected languages align with native languages
        if not cv_languages:
           language_sorted_results.append(-1)  # Assign -1 if no languages are detected
        else:
            is_sorted = any(lang in native_languages for lang in cv_languages)
            language_sorted_results.append(1 if is_sorted else 0)  # Assign 1 or 0 based on alignment
        detected_languages_list.append(cv_languages)

    # Add the results to the DataFrame
    df['language_sorted'] = language_sorted_results
    df['detected_languages'] = detected_languages_list

    # Save the updated DataFrame to a CSV file
    df.to_csv(output_csv, index=False)
    print(f"Processed results saved to {output_csv}")

# Main execution
cv_folder = r"C:\Users\Nityam\Desktop\PIRAMAL_PROJECT\CV test"  # Path to CV folder
employee_csv = "test.csv"  # Path to employee data CSV
output_csv = "processed_test.csv"  # Output file for processed data
process_employees(cv_folder, employee_csv, output_csv)
```

```python
import os
import re
from docx import Document
from nltk.sentiment import SentimentIntensityAnalyzer
import nltk

# Ensure VADER is downloaded
nltk.download('vader_lexicon')

# Initialize Sentiment Intensity Analyzer
sia = SentimentIntensityAnalyzer()

# Function to extract text from a Word document
def extract_text_from_docx(file_path):
    try:
        doc = Document(file_path)
        full_text = []
        for paragraph in doc.paragraphs:
            full_text.append(paragraph.text)
        return "\n".join(full_text)
    except Exception as e:
        return f"Error reading file {file_path}: {str(e)}"

# Function to perform sentiment analysis
def analyze_sentiment(text):
    scores = sia.polarity_scores(text)
    if scores['compound'] > 0.5:
        return "Good", scores
    elif scores['compound'] < -0.5:
        return "Bad", scores
    else:
        return "Neutral", scores

# Function to process all .docx files in a folder
def process_cvs(folder_path):
    results = {}
    good_cv_count = 0
    bad_cv_count = 0
    neutral_cv_count = 0

    for file_name in os.listdir(folder_path):
        if file_name.endswith(".docx"):
            file_path = os.path.join(folder_path, file_name)
            cv_text = extract_text_from_docx(file_path)
            
            if "Error" in cv_text:
                results[file_name] = {"Error": cv_text}
                continue
            
            sentiment, scores = analyze_sentiment(cv_text)
            results[file_name] = {
                "Extracted Text": cv_text,
                "Sentiment": sentiment,
                "Scores": scores
            }
            
            if sentiment == "Good":
                good_cv_count += 1
            elif sentiment == "Bad":
                bad_cv_count += 1
            else:
                neutral_cv_count += 1

    return results, good_cv_count, bad_cv_count, neutral_cv_count

# Specify the folder containing the CV files
folder_path = r"C:\Users\Nityam\Desktop\PIRAMAL_PROJECT\CV train"

# Process all files
all_results, good_count, bad_count, neutral_count = process_cvs(folder_path)

# Print results for each file
for file_name, result in all_results.items():
    print(f"File: {file_name}")
    if "Error" in result:
        print(f"Error: {result['Error']}")
    else:
        print(f"Sentiment: {result['Sentiment']}")
        print(f"Scores: {result['Scores']}")
    print("-" * 50)

# Summary
print("Summary:")
print(f"Good CVs: {good_count}")
print(f"Bad CVs: {bad_count}")
print(f"Neutral CVs: {neutral_count}")
```

```python
import os
import pandas as pd
from textblob import TextBlob
from docx import Document

# Function to extract text from a .docx file
def extract_text_from_docx(file_path):
    try:
        doc = Document(file_path)
        return " ".join([para.text for para in doc.paragraphs])
    except Exception as e:
        print(f"Error reading {file_path}: {e}")
        return ""

# Perform sentiment analysis on the text
def analyze_sentiment(text):
    try:
        sentiment = TextBlob(text).sentiment.polarity
        if sentiment > 0:
            return "Positive"
        elif sentiment < 0:
            return "Negative"
        else:
            return "Neutral"
    except Exception as e:
        print(f"Error analyzing sentiment: {e}")
        return "Error"

# Main function to process CVs and update the CSV
def process_cvs_and_analyze_sentiment(cv_folder, employee_csv, output_csv):
    # Read the employee data
    df = pd.read_csv(employee_csv, encoding='ISO-8859-1')

    sentiments = []
    
    for _, row in df.iterrows():
        emp_id = row['CandidateID']
        cv_filename = f"{emp_id} Resume.docx"
        cv_path = os.path.join(cv_folder, cv_filename)
        
        if not os.path.exists(cv_path):
            print(f"CV not found for CandidateID {emp_id}: {cv_filename}")
            sentiments.append("CV Not Found")
            continue
        
        # Extract text and perform sentiment analysis
        cv_text = extract_text_from_docx(cv_path)
        sentiment = analyze_sentiment(cv_text)
        sentiments.append(sentiment)
    
    # Add sentiment analysis results to the DataFrame
    df['sentiment_analysis'] = sentiments
    
    # Save the updated DataFrame to a CSV file
    df.to_csv(output_csv, index=False)
    print(f"Processed results saved to {output_csv}")

# Main execution
cv_folder = r"C:\Users\Nityam\Desktop\PIRAMAL_PROJECT\CV test"  # Path to CV folder
employee_csv = "test.csv"  # Path to employee data CSV
output_csv = "processed_test_with_sentiment.csv"  # Output file for processed data
process_cvs_and_analyze_sentiment(cv_folder, employee_csv, output_csv)
```
